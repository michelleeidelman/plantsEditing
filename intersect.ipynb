```python
#opening the bed file with the intersection as a data frame +P+Fe
df_plusP_plusFe = pd.read_csv('intersect+P+Fe.bed', sep='\t', comment='t', header=None)
header = ['chrom', 'posStart', 'posEnd', 'mutation', 'score', 'strand(gatk)', 'chrom(gtf)', 'posStart(gtf)', 'posEnd(gtf)', 'geneName', 'what', 'strand(gtf)','what2']
df_plusP_plusFe.columns = header[:len(df_plusP_plusFe.columns)]
del df_plusP_plusFe['what']
del df_plusP_plusFe['what2']
```

```python
#+P-Fe
df_plusP_minusFe = pd.read_csv('intersect+P-Fe.bed', sep='\t', comment='t', header=None)
header = ['chrom', 'posStart', 'posEnd', 'mutation', 'score', 'strand(gatk)', 'chrom(gtf)', 'posStart(gtf)', 'posEnd(gtf)', 'geneName', 'what', 'strand(gtf)','what2']
df_plusP_minusFe.columns = header[:len(df_plusP_minusFe.columns)]
del df_plusP_minusFe['what']
del df_plusP_minusFe['what2']
```

```python
#-P-Fe
df_minusP_minusFe = pd.read_csv('intersect-P-Fe.bed', sep='\t', comment='t', header=None)
header = ['chrom', 'posStart', 'posEnd', 'mutation', 'score', 'strand(gatk)', 'chrom(gtf)', 'posStart(gtf)', 'posEnd(gtf)', 'geneName', 'what', 'strand(gtf)','what2']
df_minusP_minusFe.columns = header[:len(df_minusP_minusFe.columns)]
del df_minusP_minusFe['what']
del df_minusP_minusFe['what2']
```

```python
#counting how many A2G mismatches are in every gene in the anotation (gtf)
#A2G +P+Fe:
counter_A2G_plusP_plusFe = 0
A2G_dict_plusP_plusFe={}
A2G_dict_plusP_plusFe_normalization={}
gene_name_A2G_plusP_plusFe = df_plusP_plusFe['geneName'][0]
len_Of_Genes=[]

for idx in df_plusP_plusFe.index:
    if df_plusP_plusFe['mutation'][idx]=='A_G':
        temp_gene_name_A2G_plusP_plusFe = df_plusP_plusFe['geneName'][idx]
        if temp_gene_name_A2G_plusP_plusFe==gene_name_A2G_plusP_plusFe:
            counter_A2G_plusP_plusFe += 1
            A2G_dict_plusP_plusFe[gene_name_A2G_plusP_plusFe]=counter_A2G_plusP_plusFe
        else:
            gene_name_A2G_plusP_plusFe = temp_gene_name_A2G_plusP_plusFe
            counter_A2G_plusP_plusFe = 1
            A2G_dict_plusP_plusFe[gene_name_A2G_plusP_plusFe] = counter_A2G_plusP_plusFe
            #calculating the len of every gene for normalization
            # TODO normalization and add it to the dictionary insted of the number of mismatches 
            start_pos=df_plusP_plusFe['posStart(gtf)'][idx]
            end_pos=df_plusP_plusFe['posEnd(gtf)'][idx]
            len_gene = (end_pos-start_pos) + 1
            len_Of_Genes.append(len_gene)
    

#creating 2 dataframes 1. for the gene names and the number of mismatches 2. for the gene names and the nirmalization number
A2G_name_of_genes_plusP_plusFe=A2G_dict_plusP_plusFe.items()
data_of_genes=list(A2G_name_of_genes_plusP_plusFe)
A2G_df_of_genes_plusP_plusFe=pd.DataFrame(data_of_genes)
A2G_df_of_genes_plusP_plusFe=A2G_df_of_genes_plusP_plusFe.rename(columns={0: "Gene_name", 1: "Number_of_mismatches_A2G_+P+Fe"})

A2G_df_plusP_plusFe_with_normalization= pd.DataFrame()
i = 0
for x, y in A2G_dict_plusP_plusFe.items():
    normalization=(y/len_Of_Genes[i]) * 1000
    i += 1
    temp_dict={'gene_name': x, 'normalization_number':normalization}
    A2G_df_plusP_plusFe_with_normalization=A2G_df_plusP_plusFe_with_normalization.append(temp_dict, ignore_index=True)


# x=3
# i = 0
# for x, y in A2G_dict_plusP_plusFe.items():
#     normalization=(y/len_Of_Genes[i]) * 1000
#     i += 1
#     f.write(x + ": "+str(normalization)+"\n")
```
